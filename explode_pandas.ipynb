```python
#import the lybrary
import pandas as pd
import time

#start the time to run this program (optional)
start = time.time()

#load in your spreadsheet/dataframe
df = pd.read_excel('NonSAP_Reporting_CoE_Master File (4).xlsx',skiprows=1)

#remove all the columns you don't need (optional)
df = df[['Underlying Data Elements','CleansedSourceSystem']]

#remove special characters
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^",
              "_","`","{","|","}","~","â€“","\n"]

for char in spec_chars:
    df['Underlying Data Elements'] = df['Underlying Data Elements'].str.replace(char, '')

#and lastly converty each row in the underlying data elements column to lists, 
#then explode each element in each list with the corresponding columnd (easy b/c there was only two columns))
#then call only the two columns you need (b/c this code made one extra column)
#and call head(20) which shows top 20 rows
exploded_df = pd.DataFrame(df.assign(Underlying_Data_Elements=df['Underlying Data Elements'].str.split(",")))\
.explode('Underlying_Data_Elements')\
[['Underlying_Data_Elements','CleansedSourceSystem']]

#set end time (optional)
end = time.time()

#print the time it took to run this little program (optional)
print(end - start,'seconds')

#print what the dataframe originally looked like (optional)
print(df.head())

#call exploded_df (which is the name of our dataframe at this point)
exploded_df.drop_duplicates().head(20)
```
