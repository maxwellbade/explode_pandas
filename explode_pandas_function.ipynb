```python
#explode dataframe columns function
def explode_function(df, df_explode_col, df_group_col):
    
    #drop nulls
    df = df.dropna()
    
    #load in your spreadsheet/dataframe
    df = pd.read_excel('NonSAP_Reporting_CoE_Master File (4).xlsx',skiprows=1)

    #remove all the columns you don't need (optional)
    df = df[[df_explode_col, df_group_col]]

    #remove special characters
    spec_chars = ["!",'"',"#","%","&","'","(",")",
                  "*","+","-",".","/",":",";","<",
                  "=",">","?","@","[","\\","]","^",
                  "_","`","{","|","}","~","–","\n"]

    for char in spec_chars:
        df[df_explode_col] = df[df_explode_col].str.replace(char, '')

    #and lastly converty each row in the underlying data elements column to lists, 
    #then explode each element in each list with the corresponding columnd (easy b/c there was only two columns))
    #then call only the two columns you need (b/c this code made one extra column)
    #and call head(20) which shows top 20 rows
    exploded_df = pd.DataFrame(df.assign(Exploded_Column=df[df_explode_col].str.split(",")))\
    .explode('Exploded_Column')\
    [['Exploded_Column',df_group_col]]
    
    return(exploded_df)
```

```python
#call function and exploded df
exploded_df = explode_function(df,'Underlying Data Elements', 'CleansedSourceSystem').sample(20)
exploded_df
```
